```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras import regularizers
import random
%matplotlib inline
```

```python
data = tf.keras.datasets.mnist.load_data(path="mnist.npz")
```

```python
img = data[0][0][1]
pixels = np.asarray(img)
print(pixels.shape)
plt.imshow(pixels, cmap='gray')
```

```python
train_images = data[0][0]
train_labels = data[0][1]
test_images = data[1][0]
test_labels = data[1][1]
print(train_images.shape)
print(test_images.shape)
```

```python
#sgd
model = tf.keras.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)), tf.keras.layers.Dense(128, activation='relu', kernel_regularizer = regularizers.l2(0.01)),tf.keras.layers.Dense(10)])
model.compile(optimizer="sgd", loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs = 100)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
```

```python
#using sda
print(test_acc)
```
